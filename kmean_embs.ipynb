```python
import numpy as np

def load_embs(fname):
    i2w = dict()
    embs = []
    s = 0
    V = 0
    with open(fname, 'r') as f:
        for line in f:
            p = line.strip().split()
            if len(p) == 2:
                V = int(p[0])  # Vocabulary
                s = int(p[1])  # embeddings size
            else:
                # assert len(p)== s+1
                w = "".join(p[0])
                e = [float(i) for i in p[1:]]
                if sum(e) != 0:
                    i2w[len(i2w)] = w
                    embs.append(e)
    print("Load pre-trained word vectors from file %s:" % fname)
    print("\t%d vocabulary, %d dimension" % (len(i2w), s))
    return i2w, np.array(embs)
```

```python
emb_file = "extracted_data/extracted_data/word_emb.txt"
i2w, embs = load_embs(emb_file)
```

```python
# Kmean sklearn 
from sklearn import cluster
NUM_CLUSTERS = 10
kmeans = cluster.KMeans(n_clusters=NUM_CLUSTERS)
kmeans.fit(embs)
centroids = kmeans.cluster_centers_
print ("Centroids data")
print (centroids)
```
